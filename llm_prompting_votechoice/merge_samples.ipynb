```python
prob_cols = ["p_aitc", "p_nda", "p_lf", "p_inc", "p_nota"]

# Step 1: identify and fix rows where values look like percentages (sum ~100)
pct_rows = merged["prob_sum"] > 5
print(f"Rows that look like percentages (sum > 5): {pct_rows.sum()}")
merged.loc[pct_rows, prob_cols] = merged.loc[pct_rows, prob_cols] / 100

# Recompute sum
merged["prob_sum"] = merged[prob_cols].sum(axis=1)

# Step 2: normalise all rows where will_vote==True and sum > 0
voters_mask = (merged["will_vote"] == True) & (merged["prob_sum"] > 0)
merged.loc[voters_mask, prob_cols] = (
    merged.loc[voters_mask, prob_cols]
    .div(merged.loc[voters_mask, "prob_sum"], axis=0)
)

# Step 3: set all probs to 0 for non-voters
non_voters_mask = merged["will_vote"] != True
merged.loc[non_voters_mask, prob_cols] = 0.0

# Step 4: verify
merged["prob_sum"] = merged[prob_cols].sum(axis=1)
print("\nAfter fix — probability sum stats (voters only):")
print(merged.loc[voters_mask, "prob_sum"].describe())
print(f"\nRows > 1.05: {(merged['prob_sum'] > 1.05).sum()}")
print(f"Rows < 0.95 (voters only): {(merged.loc[voters_mask, 'prob_sum'] < 0.95).sum()}")

# Step 5: save
merged.to_csv("merged_sample_v1.csv", index=False)
print("\nSaved merged_sample_v1.csv")
```

```python
# ── Fix gender typos ───────────────────────────────────────────────────────────
print("Before fix - unique gender values:")
print(sf["gender"].value_counts(dropna=False).head(20))

# Map anything that looks like Male/Female to the correct value
import re

def clean_gender(val):
    if pd.isna(val):
        return val
    v = str(val).strip().lower()
    if v in ["male", "maie", "main", "mae", "mal", "mael", "mlae", "m", "mare", "mole", "Mais"]:
        return "Male"
    if v in ["female", "fermain", "femal", "famale", "femle", "feamle", "f", "fe"]:
        return "Female"
    return val   # keep original if unrecognised — inspect manually

sf["gender"] = sf["gender"].apply(clean_gender)

print("\nAfter fix:")
print(sf["gender"].value_counts(dropna=False))

# ── Check what's still unrecognised ───────────────────────────────────────────
unrecognised = sf[~sf["gender"].isin(["Male", "Female", None])]
print(f"\nStill unrecognised: {len(unrecognised)} rows")
if len(unrecognised) > 0:
    print(unrecognised["gender"].value_counts())
```

```python
# ── Impute area_type NaN with constituency mode ────────────────────────────────
print("Before imputation:")
print(f"  NaN area_type: {sf['area_type'].isna().sum()}")
print(f"  Constituencies with NaN: {sf[sf['area_type'].isna()]['constituency_name'].nunique()}")

# Impute with constituency mode
sf["area_type"] = sf.groupby("constituency_name")["area_type"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Rur")
)

print("\nAfter imputation:")
print(sf["area_type"].value_counts(dropna=False))
print(f"  Remaining NaN: {sf['area_type'].isna().sum()}")

# ── Safety check — if any constituency had ALL NaN area_type ──────────────────
# (mode would fail — fallback to state-wide mode)
if sf["area_type"].isna().sum() > 0:
    state_mode = sf["area_type"].mode()[0]
    sf["area_type"] = sf["area_type"].fillna(state_mode)
    print(f"  Filled remaining with state-wide mode: {state_mode}")
    print(f"  Final NaN: {sf['area_type'].isna().sum()}")
```

```python
sf.to_csv("stratification_frame_cleaned.csv", index=False)
print(f"Saved stratification_frame_cleaned.csv")
print(f"  Rows       : {len(sf):,}")
print(f"  Constituencies: {sf['constituency_name'].nunique()}")
print(f"  NaN area_type : {sf['area_type'].isna().sum()}")
print(f"  NaN gender    : {sf['gender'].isna().sum()}")
```

```python
import pandas as pd

candidates = pd.read_csv('west_bengal_candidates_2026.csv')
results_2021 = pd.read_csv('wb_2021_results.csv')
results_2024 = pd.read_csv('west_bengal_assembly_with_parliament_results.csv')
sf = pd.read_csv('stratification_frame_cleaned.csv')
merged = pd.read_csv('merged_sample_v1.csv')

sets = {
    'candidates': set(candidates['constituency_name'].str.strip().str.upper()),
    'results_2021': set(results_2021['constituency_name'].str.strip().str.upper()),
    'results_2024': set(results_2024['constituency_name'].str.strip().str.upper()),
    'sf': set(sf['constituency_name'].str.strip().str.upper()),
    'merged': set(merged['constituency_name'].str.strip().str.upper()),
}

for name, s in sets.items():
    print(f'{name}: {len(s)} constituencies')

common = set.intersection(*sets.values())
print(f'\nCommon across all 5: {len(common)}')

print('\n--- Extra in each df (not in common) ---')
for name, s in sets.items():
    extra = sorted(s - common)
    print(f'\n{name} ({len(extra)} extra):')
    for c in extra:
        print(f'  {c}')
```
